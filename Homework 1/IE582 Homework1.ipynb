```r
# Getting Actual Number of Elements for Given HomeTeam Score and AwayTeam Score 
matches_premier_home_score_table = as.data.frame(table(matches_premier$match_hometeam_score))
colnames(matches_premier_home_score_table)=c('Home_Score','Num_Matches')
matches_premier_home_score_table 
matches_premier_away_score_table = as.data.frame(table(matches_premier$match_awayteam_score))
colnames(matches_premier_away_score_table)=c('Away_Score','Num_Matches')
matches_premier_away_score_table
# This will be Compared with Estimations of Poisson Model for the Dataset
matches_premier_home_score_table[order(matches_premier_home_score_table$Home_Score),]
matches_premier_home_score_char = as.data.frame(table(matches_premier$match_hometeam_score), stringsAsFactors = FALSE)
colnames(matches_premier_home_score_char)=c('Home_Score','Num_Matches')
matches_premier_home_score_char[order(matches_premier_home_score_char$Home_Score),]

matches_premier_away_score_table[order(matches_premier_away_score_table$Away_Score),]
matches_premier_away_score_char = as.data.frame(table(matches_premier$match_awayteam_score), stringsAsFactors = FALSE)
colnames(matches_premier_away_score_char)=c('Away_Score','Num_Matches')
matches_premier_away_score_char[order(matches_premier_away_score_char$Away_Score),]

# Filling the Missing Factors
j = 1
for (i in min(matches_premier$match_hometeam_score) : max(matches_premier$match_hometeam_score))
{ 
    if(matches_premier_home_score_table[j,1] == i)
    {   
        j = j+1
    }
    else
    {#matches_premier_home_score_char[j,] = matches_premier_home_score_table[i,]
        matches_premier_home_score_char[j,1] = matches_premier_home_score_char[j,1]
        matches_premier_home_score_char[j,2] = matches_premier_home_score_char[j,2]
        matches_premier_home_score_char[(j-1),1] = i
        matches_premier_home_score_char[(j-1),2] = 0
    }
}

j = 1
for (i in min(matches_premier$match_awayteam_score) : max(matches_premier$match_awayteam_score))
{ 
    if(matches_premier_away_score_table[j,1] == i)
    {   
        j = j+1
    }
    else
    {#matches_premier_away_score_char[j,] = matches_premier_away_score_table[i,]
        matches_premier_away_score_char[j,1] = matches_premier_away_score_char[j,1]
        matches_premier_away_score_char[j,2] = matches_premier_away_score_char[j,2]
        matches_premier_away_score_char[(j-1),1] = i
        matches_premier_away_score_char[(j-1),2] = 0
    }
}

matches_premier_home_score_char
matches_premier_away_score_char
```

```r
# Calculating Estimations of Poisson Model for the Dataset

# Getting the Average of HomeTeam Score and AwayTeam Score
print(paste('Number of Matches',nrow(matches_premier)))
matches_premier_home_mean = mean(matches_premier$match_hometeam_score)
matches_premier_away_mean = mean(matches_premier$match_awayteam_score)
print(paste('Home Score Mean is', matches_premier_home_mean))
print(paste('Away Score Mean is', matches_premier_away_mean))
# Calculating Poisson Estimates for HomeTeam Score
print(paste('Home score minimum is', min(matches_premier$match_hometeam_score)))
print(paste('Home score maximum is', max(matches_premier$match_hometeam_score)))
home_poisson_array = vector()
for (i in min(matches_premier$match_hometeam_score) : max(matches_premier$match_hometeam_score))
{ 
    home_poisson_array = c(home_poisson_array, nrow(matches_premier) * dpois(i, lambda=matches_premier_home_mean)) 
}
home_poisson_estimate = data.frame(min(matches_premier$match_hometeam_score) : max(matches_premier$match_hometeam_score),home_poisson_array) 
colnames(home_poisson_estimate)=c('Home_Score','Num_Matches')
home_poisson_estimate[order(home_poisson_estimate$Home_Score),]
home_poisson_estimate

# Calculating Poisson Estimates for AwayTeam Score
print(paste('Away score minimum is',min(matches_premier$match_awayteam_score)))
print(paste('Away score maximum is',max(matches_premier$match_awayteam_score)))
away_poisson_array = vector()
for (i in min(matches_premier$match_awayteam_score) : max(matches_premier$match_awayteam_score))
{ 
    away_poisson_array = c(away_poisson_array, nrow(matches_premier) * dpois(i, lambda=matches_premier_away_mean)) 
}
away_poisson_estimate = data.frame(min(matches_premier$match_awayteam_score) : max(matches_premier$match_awayteam_score),away_poisson_array) 
colnames(away_poisson_estimate)=c('Away_Score','Num_Matches')
away_poisson_estimate

#ppois(max(matches_premier$match_hometeam_score), matches_premier_home_mean)
#home_poisson_simulation_table = as.data.frame(table(rpois(nrow(matches_premier), matches_premier_home_mean)))
#home_poisson_simulation_table

dpois(0, lambda=1.228)*618
```

```r
#Merging Bets and Matches Tables
bets_matches = merge(x=bets,y=matches,by='match_id',all.x=TRUE)
# Filtering Finished Premier League Matches from the Merged Table.
bets_matches_premier = bets_matches[bets_matches$league_id==148 & bets_matches$match_status=='Finished', bets_matches$epoch != NA, , bets_matches$odd_bookmakers != NA]
```

```r
# Getting List of Odd_bookmakers 
odd_bookmakers = data.frame(unique(bets_matches_premier$odd_bookmakers))
# Selecting Odd_bookmakers randomly with a Given Seed
set.seed(2006703609)
sample <- sample(1:nrow(odd_bookmakers), 4)
sample
odd_bookmakers[sample[1],1]
odd_bookmakers[sample[2],1]
odd_bookmakers[sample[3],1]
odd_bookmakers[sample[4],1]
```

```r
# bwin, BetOnline, BetGun, Betway were selected
bets_matches_premier_bwin=bets_matches_premier[bets_matches_premier$odd_bookmakers =='bwin',]
# Getting odds
bets_matches_premier_bwin=bets_matches_premier_bwin[bets_matches_premier_bwin$variable=='odd_1'  | bets_matches_premier_bwin$variable=='odd_x' | bets_matches_premier_bwin$variable=='odd_2' ,]

# TASK 2.1
# Calculating Probabilities ,  P(home win), P(tie) and P(away win) by P(x) = 1/odd
bets_matches_premier_bwin$implied_probability=1/(as.numeric(as.character(bets_matches_premier_bwin$value)))
bet_probs_bwin = bets_matches_premier_bwin  %>% group_by(match_id, odd_epoch) %>% summarise(sum(implied_probability))
colnames(bet_probs_bwin)[3]='probability_total'
# TASK 2.2
# Calculating Normalized Probabilities
bets_matches_premier_bwin=merge(x=bets_matches_premier_bwin,y=bet_probs_bwin,by=c('match_id','odd_epoch'),all.x=TRUE)
bets_matches_premier_bwin['norm_prob']=bets_matches_premier_bwin['implied_probability'] * (1/bets_matches_premier_bwin['probability_total'])
bets_matches_premier_bwin_simp = bets_matches_premier_bwin[, c('match_id','odd_epoch','variable','norm_prob')]
bets_matches_premier_bwin_simp = bets_matches_premier_bwin_simp[complete.cases(bets_matches_premier_bwin_simp), ]
bets_matches_premier_bwin_odds = dcast(bets_matches_premier_bwin_simp, match_id + odd_epoch ~ variable, value.var = "norm_prob")
```

```r
# TASK 2.3
bets_matches_premier_bwin_odds[,"Home_Lead"] = bets_matches_premier_bwin_odds[,"odd_1"] - bets_matches_premier_bwin_odds[,"odd_2"]  
bets_matches_premier_bwin_odds

bets_matches_premier_bwin_diff=bets_matches_premier_bwin_odds[,c('match_id','odd_epoch','odd_x','Home_Lead')]
plot(bets_matches_premier_bwin_diff[,4],bets_matches_premier_bwin_diff[,3],xlab = ' P(home win) – P(away win)',ylab='P (tie)', main='Predictions of bwin')
```

```r
# Adding Bins
bets_matches_premier_bwin_diff["Bin"] = 0
bets_matches_premier_bwin_diff$Bin <- cut(bets_matches_premier_bwin_diff$Home_Lead, breaks = c(-1, -0.8, -0.6, -0.4, -0.2, 0, 0.2, 0.4, 0.6, 0.8, 1), 
labels = seq(1,10))
bets_matches_premier_bwin_diff
```

```r
# Adding Bins and Checking if the Match is Actually Drawn
bets_matches_premier_bwin_large=merge(x=bets_matches_premier_bwin_diff,y=bets_matches_premier_bwin[,c('match_id','odd_epoch','match_hometeam_score','match_awayteam_score','norm_prob')],by=c('match_id','odd_epoch'),all.x=TRUE)
bets_matches_premier_bwin_draw=bets_matches_premier_bwin_large[,c('match_id','odd_epoch','match_hometeam_score','match_awayteam_score','norm_prob','Home_Lead','Bin')]
bets_matches_premier_bwin_draw['is_drawn'] = 0
bets_matches_premier_bwin_draw['is_drawn'] <- ifelse(bets_matches_premier_bwin$match_hometeam_score == bets_matches_premier_bwin$match_awayteam_score,1,0)
bets_matches_premier_bwin_draw
```

```r
bets_matches_premier_bwin_draw_prob =bets_matches_premier_bwin_draw %>% group_by(Bin) %>% summarise(count=n(),sum(is_drawn))
colnames(bets_matches_premier_bwin_draw_prob)[3]='draw'
bets_matches_premier_bwin_draw_prob$draw_probability=bets_matches_premier_bwin_draw_prob$draw/bets_matches_premier_bwin_draw_prob$count
bets_matches_premier_bwin_draw_prob
bets_matches_premier_bwin_draw_prob=bets_matches_premier_bwin_draw_prob[,c(1,4)]
```

```r
# Task 1.a

# Applying PCA
musk1_components = musk1_data[,3:length(musk1_data)]
musk1_pca=princomp(scale(musk1_components))
summary(musk1_pca)
```

```r
# Plotting PCA 
pca_musk_variance = musk1_pca$sdev^2
pca_musk_percentage = round(pca_musk_variance/sum(pca_musk_variance)*100,2)
plot(pca_musk_percentage, main = "Comp Plot",
     xlab = "Principal Component",
     ylab = "% of Variation", 
     type = "b",
     col="blue")
plot(cumsum(pca_musk_percentage), main = "Cumulative",
     xlab = "Principal Component",
     ylab = "Cum. % of Variance",
     type = "b",
     col="blue")
```

# Task 2.1

```r
# Reading the image
own_img=readJPEG("C:/Users/emree/IE582/Emre.jpg",native=FALSE)
# resize function in imager package did not work, therefore uploaded image is 256 x 256
str(own_img)
```

```r
# Displaying the image in 
ximage = 256
yimage = 256
plot(1:ximage, type='n',main="Emre Photo 256x256")
rasterImage(own_img,0,0,ximage,yimage,angle = 0,interpolate = F) 
```

# Task 2.2

```r
# Adding Noise 

# Getting Min and Max Pixel Values
min_pixel_value = min(own_img)
max_pixel_value = max(own_img)/10
min_pixel_value
max_pixel_value

# Adding Uniform Noise by Noise Matrix
noise_matrix <-array(0,dim = c(ximage,yimage,3))
for(i in 1:ximage){
    for(j in 1:yimage){
        for(k in 1:3){
          uniform_random<- runif(1, min = min_pixel_value, max = max_pixel_value)
          noise_matrix[i,j,k]<-uniform_random
        }
    }
}
noisy_own_img <-array(0,dim = c(ximage,yimage,3))
for(i in 1:ximage){
    for(j in 1:yimage){
        for(k in 1:3){            
          noisy_own_img[i,j,k]<-own_img[i,j,k] + noise_matrix[i,j,k]
        }
    }
} 
# Normalizing After Adding Noise Matrix
noisy_own_img <-noisy_own_img/max(noisy_own_img)
```

```r
# Task 2.2.a

# Displaying the noisy image 
plot(1:ximage, type='n',main="Noisy Emre Emil 256x256")
rasterImage(noisy_own_img,0,0,ximage,yimage,angle = 0,interpolate = F) 
```

```r
# Task 2.2.b

# Creating Red Green Blue Channel Images

noisy_red_img<-noisy_own_img
noisy_red_img[,,2]=0 # green = 0
noisy_red_img[,,3]=0 # blue = 0

noisy_green_img<-noisy_own_img
noisy_green_img[,,1]=0 # red = 0
noisy_green_img[,,3]=0 # blue = 0

noisy_blue_img<-noisy_own_img
noisy_blue_img[,,1]=0 # red = 0
noisy_blue_img[,,2]=0 # green = 0


# Displaying each channel separately using image function on a single plot. 

par(mfrow=c(1,3))
plot(1:ximage ,type='n',main = "Noisy Red")
rasterImage(noisy_red_img,0,0,ximage,yimage,angle = 0,interpolate = F)
plot(1:ximage ,type='n',main = "Noisy Green")
rasterImage(noisy_green_img,0,0,ximage,yimage,angle = 0,interpolate = F)
plot(1:ximage ,type='n',main = "Noisy Blue")
rasterImage(noisy_blue_img,0,0,ximage,yimage,angle = 0,interpolate = F)
```

# Task 2.3

```r
# Making Greyscale by Adding RGB values and Normalizing
noisy_grey_image <- noisy_own_img[,,1]+noisy_own_img[,,2]+ noisy_own_img[,,3]
noisy_grey_image <- noisy_grey_image / max(noisy_grey_image)
```

```r
# Plotting Greyscale Image
plot(1:ximage,main = "Noisy Grey Image", xlab = "", ylab = "", axes = T, col = 0)
rasterImage(noisy_grey_image,0,0,ximage,yimage,angle = 0,interpolate = F)
```

```r
# Getting Patches
# number of patches are (ximage - xpatch + 1) x (yimage - ypatch + 1)  
# = (256 - 25 + 1) x (256 - 25 + 1) = 232 x 232
xpatch = 25
ypatch = 25

patch_size <- rep(NA,xpatch*ypatch)
patch_matrix <- rep(list(patch_size), (ximage-xpatch+1)*(yimage-ypatch+1))

k <- 1
for (i in 1:(ximage-xpatch+1)) {
  for (j in 1:(yimage-ypatch+1)) {
    patch_matrix [[k]] <- as.vector(noisy_grey_image[(i):(i+xpatch-1),(j):(j+ypatch-1)])
    k <- k+1
  }
}
```

```r
# Task 2.3.a

#Converting Matrix to Data Table
patch_table <- as.data.table(matrix(unlist(patch_matrix), ncol = 625, byrow = TRUE))

#Applying PCA to this data matrix 
pca_grey_image=princomp(patch_table)
plot(pca_grey_image)
```

```r
summary(pca_grey_image)
```

```r
pca_img_variance = pca_grey_image$sdev^2
pca_img_percentage = round(pca_img_variance/sum(pca_img_variance)*100,2)
plot(pca_img_percentage, main = "Comp Plot",
     xlab = "Principal Component",
     ylab = "% of Variation", 
     type = "b",
     col="blue")
plot(cumsum(pca_img_percentage), main = "Cumulative",
     xlab = "Principal Component",
     ylab = "Cum. % of Variance",
     type = "b",
     col="blue")
```

```r
# There are 625 components 25 x 25
# Comp 1 explains more than 85% of the vairance, 
# Comp 1 + Comp 2 + Comp 3 xplains more than 93% of the vairance,
# Other components do not explain much of the variance. They might be considered to explain noise, deformations, etc..
```

```r
str(pca_grey_image$scores)
```

```r
# Task 2.3.b

# Reconstructing the Image on 3 Principal Component 
comp1_image=matrix(0,(ximage-xpatch+1),(yimage-ypatch+1))
k=1
for(i in 1:(ximage-xpatch+1))
{
  comp1_image[i,]=pca_grey_image$scores[k:((ximage-xpatch+1)+k-1),1]
  k= k + (ximage-xpatch+1)
}

comp2_image=matrix(0,(ximage-xpatch+1),(yimage-ypatch+1))
k=1
for(i in 1:(ximage-xpatch+1))
{
  comp2_image[i,]=pca_grey_image$scores[k:((ximage-xpatch+1)+k-1),2]
  k= k + (ximage-xpatch+1)
}

comp3_image=matrix(0,(ximage-xpatch+1),(yimage-ypatch+1))
k=1
for(i in 1:(ximage-xpatch+1))
{
  comp3_image[i,]=pca_grey_image$scores[k:((ximage-xpatch+1)+k-1),3]
  k= k + (ximage-xpatch+1)
}
```

```r
# Plotting Image of Comp 1
plot(c(0,1),c(0,1),t='n')
# Normalized Version
rasterImage((comp1_image-min(comp1_image))/(max(comp1_image)-min(comp1_image)),0,0,1,1)
```

```r
# Plotting Image of Comp 2
plot(c(0,1),c(0,1),t='n')
# Normalized Version
rasterImage((comp2_image-min(comp2_image))/(max(comp2_image)-min(comp2_image)),0,0,1,1)
```

```r
# Plotting Image of Comp 3
plot(c(0,1),c(0,1),t='n')
# Normalized Version
rasterImage((comp3_image-min(comp3_image))/(max(comp3_image)-min(comp3_image)),0,0,1,1)
```

```r
# Component 1 seems to best identify the original greyscale image. 
# The other components explain some of the variance. But are more shallow and vague
```

```r
# Task 2.3.c

# Constructing Eigenvectors
eigenvector1 <- matrix((pca_grey_image$loadings[,1 ]-min(pca_grey_image$loadings[,1 ]))/(max(pca_grey_image$loadings[,1 ])-min(pca_grey_image$loadings[,1 ])),xpatch,byrow=TRUE)
eigenvector2 <- matrix((pca_grey_image$loadings[,2 ]-min(pca_grey_image$loadings[,2 ]))/(max(pca_grey_image$loadings[,2 ])-min(pca_grey_image$loadings[,2 ])),xpatch,byrow=TRUE)
eigenvector3 <- matrix((pca_grey_image$loadings[,3 ]-min(pca_grey_image$loadings[,3 ]))/(max(pca_grey_image$loadings[,3 ])-min(pca_grey_image$loadings[,3 ])),xpatch,byrow=TRUE)
```

```r
# Displaying Eigenvector 1
plot(1:25,1:25, ann = TRUE,col=0, main = "EigenVector1", xlab = "", ylab = "")
rasterImage(matrix(eigenvector1, xpatch, byrow=TRUE),1,1,xpatch,ypatch)
```

```r
plot(1:25,1:25, ann = TRUE,col=0, main = "EigenVector2", xlab = "", ylab = "")
rasterImage(matrix(eigenvector2, xpatch, byrow=TRUE),1,1,xpatch,ypatch)
```

```r
plot(1:25,1:25, ann = TRUE,col=0, main = "EigenVector3", xlab = "", ylab = "")
rasterImage(matrix(eigenvector3, xpatch, byrow=TRUE),1,1,xpatch,ypatch)
```
